```python
import geopandas
gdf = geopandas.read_file("data/geo_raw/krs/krs.shp")
gdf["krs_code"] = gdf.krs_code.astype("int")
gdf
```

```python
geo_kreise = staedte.merge(gdf, how="left", on="krs_code")
geo_kreise
```
